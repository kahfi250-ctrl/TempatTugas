Apa itu algoritma?

1. algoritma adalah urutan langkah-langkah logis yang digunakan untuk menyelesaikan masalah.
2. setiap langkah harus jelas dan dapat dieksekusi.
3. Algoritma adalah dasar dari pemrograman komputer, membantu memecahkan masalah secara efisien.

Definisi program/pemrograman
- adalah kumpulan instruksi-intruksi tersendiri yang biasanya disebut dengan source code yang dibuat oleh programmer ( pembuat program )

Paradigma pemrograman
1. pemrograman prosedural
- berdasarkan urutan-urutan, sekuensial
- program adalah suatu rangkaian prosedur untuk memanipulasi data. Prosedur merupakan kumpulan instruksi yang dikerjakan secara berurutan.
- harus mengingat prosedur mana yang sudah dipanggil dan apa yang sudah diubah

2. pemrograman fungsional
- berdasarkan teori
- fungsi merupakan dasar utama program

3. pemrograman terstruktur
- secara berurutan dan terstruktur
- program dapat dibagi-bagi menjadi prosedur dan fungsi
- contoh : PASCAL dan C

4. pemrograman modular
- pemgrograman ini membentuk banyak modul
- modul merupakan kumpulan dari prosedur dan fungsi yang berdiri sendiri
- sebuah program dapat merupakan kumpulan modul-modul
- contoh : MODULA-2 ATAU ADA

5. pemrograman berorientasi obyek
- pemrograman berdasarkan prinsip obyek, dimana obyek memiliki data/variable/property yang dapat dimanipulasi
- contoh : C++, Object pascal, dan Java

6. pemrograman berorientasi fungsi
- pemrograman ini berfokus pada suatu fungsi tertentu saja. Sangat tergantung pada tujuan pembuatan bahasa pemrograman ini.
- contoh : SQL(Structured Query Language), html, xml, dan lain-lain

7. pemrograman deklaratif
- pemrograman ini mendeskripsikan suatu masalah dengan pernyataan daripada memecahkan masalah dengan memimplementasikan algoritma.
- contoh : PROLOG

pemakai ( brainware )

1. asal usul istilah "algoritma"
- kata "algoritma" berasal dari nama seorang ahli matematika persia, Abu Abdullah muhammad ibn musa al-khawarizmi, yang hidup pada abad-9
- nama Al-khawarizmi mengalami perubahan fonetis di eropa pada abad pertengahan menjadi "algoritmi", dan disitulah istilah "algoritma" muncul.

2. algoritman di dunia kuno
- walaupun istilah "algoritma" baru muncul pada abad-9, konsep penyelesaian masalah melalui langkah-langkah sistematis sudah ada sejak zaman kuno
- Tabel perhitungan banilona (sekitar 1600 SM) menunjukan teknik-teknik komputasi dasar.
- Euclid (abad ke-3 SM) menyusun algoritma yang terkenal, Algoritma Euclide, untuk menemukan pembagian terbesar dua bilangan.
- Archimedes juga menggunakan algoritma dalam karyanya tentang perhitungan volume dan luas

3. algoritma dalam matematika klasik
- pada abad pertengahan dan renaisans,  perkembangan agoritma berfokus pad aaritmatika, terutama di bidang perhintungan manual.
- John Napier, matematikawan skotlandia pada awal abad ke-17, memperkenalkan logaritma, suatu konsep matematis yang kemudian diformalasikan sebagai algoritma untuk mempermudah perhitungan multiplikasi dan pembagian.
- blaise pascal dan gottfried wielhelm leibniz di abad ke-17 juga mengembangkan mesin mekanik berbasis algoritma untuk melakukan komputasi aritmatika

4. revolusi algoritma dan komputer
- revolusi algoritma dimulai dengan penemuan mesin komputasi modern. Charles babbage, seorang matematikawan inggris di abad ke-19, merancang mesi analitik, yang dianggap sebagai komputer mekanik pertama
- Ada Lovelace, seorang ahli matematika inggris dan kaloborator babbage, dianggap sebagai programmer pertama di dunia

5. algoritma di era komputasi modern
- pada abad ke-20, perkembangan ilmu komputer menciptakan kebutuhan besar akan algoritma yang lebih kompleks
- Alan turing, yang dikenal sebagai bapak ilmu komputer modern, mengembangkan model komputasi teoretis yang disebut mesin turing pada tahun 1936
- john von neumann juga memberikan kontribusi besar dalam pengembangan komputer modern dengan merancang arsitektur komputasi yang mendukung eksekusi algoritma secara efesien

6. perkembangan algoritma di dunia digital
- seiring berkembangnya teknologi komputer, algoritmma menjadi pusat dari berbagai aplikasi komptasi modern
- algoritma sorting seperi buuble sort, quick sort, dan marge sort mulai dikembangkan untuk mengatur data secara efisien
- pada 1950-an dan 1960-an, donald knuth menerbitkan buku "The art of computer programming",
- algoritma kriptografi, seperti RSA (Rivest-Shamir-Adleman), dikembangkan pada 1970-an untuk mengamankan data didunia digital, yang menjadi dasar dari komunikasi yang aman saat ini

7. algoritma di era kecerdasan buatan
- pada abad ke-21, algoritma memegang perananan penting dalam perkembangan kecerdasan buatan ( AI ), machine learning, dan big data.
- algoritma neural networks daan deep learning memungkinkan komputer untuk "belajar" dari data dan membuat keputusan yang kompleks.
- penerapan algoritma pada searce engine ( google ), media sosial, hingga otomasi di berbagai industri membuat algoritma menjadi bagian tak terpisahkan dari kehidupan modern.

8. masa depan algoritma
- di masa depan, algoritma terus berkembang dengan peningktan dalam komputasi kuantum, pengenalan pola, dan optimasi sistem kompleks.
- algoritma tidak hanya berperan dalam komputerisasi tetapi juga dalam memecahkan masalah dibidang lain seperti medis, keuangan, dan robotik
